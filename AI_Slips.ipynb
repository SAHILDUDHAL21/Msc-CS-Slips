# BFS Travelsal & BFS Search

```python
from collections import deque
```

```python
g={
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':['G'],
    'E':['H'],
    'F':['I'],
    'G':[],
    'H':[],
    'I':[]
}
```

```python
def bfs_c(s):
    visited=[]
    q=deque([s])

    while q:
        node =q.popleft()
        visited.append(node)

        for n in g[node]:
            q.append(n)

    return visited
```

```python
def bfs(s,t):
    visited=[]
    q=deque([s])

    while q:
        node =q.popleft()
        visited.append(node)

        if node==t:
            return visited

        for n in g[node]:
            q.append(n)

    return visited
```

```python
print("BFS Complete ",bfs_c('A'))
```

```python
print("BFS Search F ",bfs('A','F'))
```

# DFS Travelsal & Search

```python
def dfs_c(node,visited):
    visited.append(node)
    for n in g[node]:
        dfs_c(n,visited)
    return visited
```

```python
def dfs(node,t,visited):
    visited.append(node)

    if node==t:
        return True

    for n in g[node]:
        if dfs(n,t,visited):
            return visited

    return False
```

```python
print("DFS Complete ",dfs_c('A',[]))
```

```python
print("DFS Search ",dfs('A','H',[]))
```

# DLS

```python
def dls(node,t,visited,l):
    visited.append(node)
        
    if node==t:
        return True

    if(l==0):
        return False

    for n in g[node]:
        if dls(n,t,visited,l-1):
            return visited

    return False
```

```python
print("dls Search ",dls('A','H',[],3))
```

```python
def dfs_1(s,t):
    visited=[]
    q=[s]

    while q:
        node =q.pop()
        visited.append(node)

        if node==t:
            return visited

        for n in g[node]:
            q.append(n)

    return visited
```

```python
print("dls Search F ",dfs_1('A','H'))
```

# Water Jug

```python
from collections import deque
```

```python
def w_j(j1,j2,t):
    q=deque([((0,0),[])])
    seen=set([(0,0)])
    while q:
        (a,b),p=q.popleft()
        p=p+[(a,b)]
        if a==t or b==t :
            for x,y in p:
                print(f"Jug1: {x}L, Jug2 {y}L")                
            print("Done!")
            return
        
        for x,y in [(j1,b),(a,j2),(0,b),(a,0),(a- min(a,j2-b),b+min(a,j2-b)),(a+min(b,j1-a),b-min(b,j1-a))]:
            if (x,y) not in seen:
                seen.add((x,y))
                q.append(((x,y),p))  
```

# UCS

```python
import heapq
```

```python
def ucs(g,s,goal):
    f=[(0,s,[s])]
    visited=set()
    while f:
        c,node,p=heapq.heappop(f)
        if node in visited:
            continue
        visited.add(node)
        if node==goal:
            return c,p

        for n, edge in g[node]:
            if n not in visited:
                heapq.heappush(f,(c+edge,n,p+[n]))
    return float("int"),[]
```

```python
g1={
    'A':[('B',2),('C',3)],
    'B':[('D',3),('E',1)],
    'C':[('F',2)],
    'D':[('G',4)],
    'E':[('H',2)],
    'F':[('I',1)],
    'G':[],
    'H':[],
    'I':[]
}
```

```python
print("UCS A->I  ",ucs(g1,'A','I'))
```

# A*

```python
import heapq

def astar(graph, heuristics, start, goal):
    open_list = []
    heapq.heappush(open_list, (heuristics[start], 0, start, [start]))  
    # (f = g+h, g, node, path) 

    visited = set()

    while open_list:
        f, g, node, path = heapq.heappop(open_list)

        if node == goal:
            return path

        if node in visited:
            continue
        visited.add(node)

        for neighbor, cost in graph[node].items():
            new_g = g + cost
            new_f = new_g + heuristics[neighbor]
            heapq.heappush(open_list, (new_f, new_g, neighbor, path + [neighbor]))

    return None


graph = {
    'A': {'B': 1, 'C': 3},
    'B': {'D': 3, 'E': 1},
    'C': {'F': 5},
    'D': {'G': 2},
    'E': {'G': 2},
    'F': {'G': 2},
    'G': {}
}

heuristics = {
    'A': 7, 'B': 6, 'C': 5,
    'D': 4, 'E': 2, 'F': 1, 'G': 0
}

print("Optimal Path:", astar(graph, heuristics, 'A', 'G'))
```

# Bidirectional Search

```python
from collections import deque

def bidirectional_search(g, s, t):
    if s == t: return [s]
    qs, qt = deque([s]), deque([t])
    vs, vt = {s: None}, {t: None}

    while qs and qt:
        u = qs.popleft()
        for v in g[u]:
            if v not in vs:
                vs[v] = u
                qs.append(v)
                if v in vt: return path(vs, vt, v)

        u = qt.popleft()
        for v in g[u]:
            if v not in vt:
                vt[v] = u
                qt.append(v)
                if v in vs: return path(vs, vt, v)
    return None

def path(vs, vt, meet):
    p1, p2 = [], []
    x = meet
    while x is not None: p1.append(x); x = vs[x]
    x = vt[meet]
    while x is not None: p2.append(x); x = vt[x]
    return p1[::-1] + p2

# Example
graph = {
    'A':['B','C'], 'B':['A','D'], 'C':['A','F'],
    'D':['B'], 'F':['C']
}

print(bidirectional_search(graph, 'A', 'F'))
```

```python
import math

def win(b):
    W = [(0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6)]
    for a,b1,c in W:
        if b[a] == b[b1] == b[c] != '-': return b[a]
    return None

def minimax(b, mx):
    w = win(b)
    if w == 'X': return 1
    if w == 'O': return -1
    if '-' not in b: return 0

    best = -math.inf if mx else math.inf
    for i in range(9):
        if b[i] == '-':
            b[i] = 'X' if mx else 'O'
            val = minimax(b, not mx)
            b[i] = '-'
            best = max(best, val) if mx else min(best, val)
    return best

def best_move(b):
    best, move = -math.inf, -1
    for i in range(9):
        if b[i] == '-':
            b[i] = 'X'
            val = minimax(b, False)
            b[i] = '-'
            if val > best: best, move = val, i
    return move

board = ['-']*9
print("You: O | Computer: X")

while True:
    print(board[0:3], board[3:6], board[6:9])
    p = int(input("0-8: "))
    board[p] = 'O'

    if win(board): print("You win!"); break
    if '-' not in board: print("Draw!"); break

    c = best_move(board)
    board[c] = 'X'

    if win(board): print("Computer wins!"); break
```
